```python
# Dictionary to store the characters
characters = {}
# List of all node_numbers for the characters
char_vals = []

# f3 has the merged dict of all words
# We try to get the words that can be characters in the novel. The main features for these words are:
#          i. Only first letter is capitalised
#         ii. Total no of chars is more than 1

# To check these, we first remove all punctuations from each word.
# then we check if it is a valid english word.
# The words that are valid english words are skipped
f3 = open('output-10/dicts_and_encoded_texts/dict_merged.txt','r')
for line in f3:
    line1 = line.split("\t")
    
    word = line1[0]
    node_value = int(line1[1][0:len(line1[1])-1])
    
    # remove punctuation from each word
    word = remove_punct(word)
    print(line1[0],'-----------',word)
    # check if only the 1st char is capitalised. 
    # length of word is more than 1
    if((len(word) > 1) and (word[0].isupper())):
                    
        # check if it is an english word
        if(d.check(string.lower(word))):
            continue
        
        characters[node_value] = word
        char_vals.append(node_value)

print(len(characters))
print("char: ",characters)
```

```python
# Open the graph's edge list
f = open("output-10/graph/encoded_edges_count_window_size_10.txt","r")

f1 = f.readlines()

source = []
destination = []
weights = []

# Read each list
for line in f1:
    line = line.split("\t")
    #print(line)
    
    # Get the edge details
    # If both nodes are characters, then append to lists
    if((int(line[0]) in char_vals) and (int(line[1]) in char_vals)):
        source.append(int(line[0]))
        destination.append(int(line[1]))
        weight = int(line[2][:len(line[2])-1])
        weights.append(weight)
print(len(set(source)) , len(source))
print(len(set(destination)) , len(destination))
#print(type(source),type(destination))
```
